```python
df = pd.read_excel('datasets/Canada.xlsx',
                   sheet_name = 'Canada by Citizenship',
                   engine = 'openpyxl',
                   skiprows = range(20),
                   skipfooter = 2
                  )
df.head(2)
```

### decpriptional dataset

This dataset explains the Citenzship of immigrants that arrive into Canada

```python
df.columns
```

```python
df.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV', 'DevName'],
        axis = 1,
        inplace = True)
df.columns
```

```python
df.drop(['Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],
        axis = 1,
        inplace = True)
df.columns
```

```python
df.rename(columns = { 'OdName' : 'Country' ,
                     'AreaName' : 'Continent',
                     'RegName' : 'Region'},
          inplace = True
         )
df.head()
```

```python
df.columns = list(map(str,df.columns))
df.columns
```
