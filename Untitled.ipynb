# Base de datos de Medallistas olimpicos

```python
df = pd.read_csv('Medals_Standing.csv')
df.head()
```

```python
x = df.columns.tolist()
x.sort()
x
```

# Descripcion de los datos

`NOC:` Pais del atleta, str  

`Gold:` Medalla de 1 lugar, int

`Silver:` Medalla de 2 lugar, int

`Bronze:` Medalla de 3 lugar, int

`Total:` Total de medallas, int

```python
df.set_index('NOC', inplace = True)
df.head()
```

`Dato interesante:` Noruega fue el país con mayor cantidad de medallas

# Gráfica de comparación de las medallas por país

```python
mpl.style.use('ggplot')
```

```python
df.loc['Norway':].plot()
```
